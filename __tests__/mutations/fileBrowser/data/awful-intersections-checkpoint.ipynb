# Awful Intersections

This demo notebook uses the Overpass API to find "awful intersections" in a city.

It was inspired by code created by [j6k4m8](https://github.com/j6k4m8/) and art by [Barely Maps](https://www.etsy.com/shop/BarelyMaps).

The notebook uses the density, proximity, and location of traffic lights in a city to estimate the awfulness of an intersection. 
To do this, we compute the distance between all street lights in a city, cluster, and perform some post-processing to pick out awful intersections.

You can try generating the awful intersections in your city by setting the `location_name` variable. There are
other interesting things to try like changing the thresholding distances, order and methods for sorting the clusters,
and settings for plotting.

````python
%matplotlib inline
from IPython.core.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np
import osmnx as ox
import glob
import os
import shutil
import json
from jinja2 import Template
from PIL import Image
import imgkit
import overpy
from sklearn.cluster import DBSCAN

from awful.distance import get_haversine_distances
from awful.viz import clean_name, render_html_template
````

### Set the location to query
This `location_name` variable sets the region queried. You can play with this setting visually using (overpass turbo)[https://overpass-turbo.eu/#]. The query to try is:

```
[out:json][timeout:25];
{{geocodeArea:"<ENTER LOCATION NAME HERE>"}}->.searchArea;
(
  node["highway"="traffic_signals"](area.searchArea);
);
out body;
>;
out skel qt;
```

````python
# Set the location variable
location_name = "District of Columbia"
````

````python
# You can easily access the input, code, and output directries using envirnment variables that are set in every LabBook
# LB_INPUT -> /input
# LB_OUTPUT -> /output
# LB_CODE -> /code
input_dir = os.environ["LB_INPUT"]
output_dir = os.environ["LB_OUTPUT"]
code_dir = os.environ["LB_CODE"]
````

````python
api = overpy.Overpass()
TRAFFIC_SIGNAL_QUERY = f"""
[out:json]
[timeout:2500]
;
area["name"="{location_name}"]->.searchArea;
(
  node["highway"="traffic_signals"](area.searchArea);
);
out;
>;
out skel qt;
"""
````

````python
response = api.query(TRAFFIC_SIGNAL_QUERY)
traffic_lights = response.nodes
````

````python
# Ge matrix of lat long
X = np.array([[float(tl.lat), float(tl.lon)] for tl in traffic_lights])
print(f"Found {len(X)} traffic lights in {location_name}")
````

````python
traffic_light_distances = get_haversine_distances(X, X)
````

````python
plt.figure(figsize = (10,10))
plt.imshow(traffic_light_distances)
_ = plt.title(f"{location_name} traffic light distances")
_ = plt.xlabel("Traffic Light Index")
_ = plt.ylabel("Traffic Light Index")
````

````python
# Threshold distance matrices to only keep lights that are "close" to each other
distance_threshold_ft = 500
distance_threshold_mi = distance_threshold_ft/5280

potentials = traffic_light_distances < distance_threshold_mi 
````

````python
# Plot thresholded distance matrices
plt.figure(figsize = (10,10))
plt.imshow(potentials)
_ = plt.title(f"{location_name} thresholded traffic light distances")
_ = plt.xlabel("Traffic Light Index")
_ = plt.ylabel("Traffic Light Index")
````

````python
# Remove diagonal, since a traffic light will always be within distance_threshold_ft from itself!
np.fill_diagonal(potentials, False)

# Extract upper triangle only
potentials = np.triu(potentials, -1)

plt.figure(figsize = (10,10))
plt.imshow(potentials)
_ = plt.title(f"{location_name} thresholded traffic light distances")
_ = plt.xlabel("Traffic Light Index")
_ = plt.ylabel("Traffic Light Index")
````

````python
candidates_lat_long = [X[r[0]] for r in np.array(np.where(potentials > 0)).T]
candidates_lat_long = np.array(candidates_lat_long)
print(f"Found {len(candidates_lat_long)} traffic lights in within {distance_threshold_ft}ft of another light")
````

````python
# Cluster the thresholded distance matrix, to find clusters of lights close to each other
# These clusters are a proxy for finding awful intersections
clusters = DBSCAN(eps=0.001, min_samples=4).fit(candidates_lat_long)

plt.figure(figsize = (10,10))
plt.scatter(*candidates_lat_long.T, c=clusters.labels_)
_ = plt.title(f"{location_name} traffic light clusters")
_ = plt.xlabel("Latitude")
_ = plt.ylabel("Longitude")
````

````python
complex_intersections = [[] for _ in range(np.max(clusters.labels_))]

for i in range(len(candidates_lat_long)):
    if clusters.labels_[i] > -1:
        light = candidates_lat_long[i]
        complex_intersections[clusters.labels_[i]-1].append(light)

complex_intersections = np.array([np.array(c) for c in complex_intersections])
````

````python
# Get up to 25 intersections with lots of lights (i.e. 25 clusters)
num_lights_per_cluster = [x.shape[0] for x in complex_intersections]

sort_idx = np.array(num_lights_per_cluster).argsort()
sorted_complex_intersections = complex_intersections[sort_idx[::-1]]
sorted_complex_intersections = sorted_complex_intersections[:25]
````

````python
# Then try sorting again on the variance, making more lights spread around more awful
cluster_variance = [np.var(x) for x in sorted_complex_intersections]
sort_idx = np.array(cluster_variance).argsort()
sorted_complex_intersections = sorted_complex_intersections[sort_idx[::-1]]
````

````python
# Get centroids of the clusters
intersection_centroids = [[np.mean(c[:, 0]), np.mean(c[:, 1])] for c in sorted_complex_intersections]
````

````python
# Prep output
result_dir = os.path.join(output_dir, "awful_intersections")
img_dir = os.path.join(result_dir, "images")

# Clean out old results if they are present
if os.path.exists(result_dir):
    shutil.rmtree(result_dir)

# Make output dirs
if not os.path.exists(result_dir):
    os.makedirs(result_dir)
if not os.path.exists(img_dir):
    os.makedirs(img_dir)
````

````python
# Generate images and links to google maps for each intersection
plot_radius_meters = ((distance_threshold_ft/ 2) * 0.3048) + 50
fnames = []

cwd = os.getcwd()
os.chdir(result_dir)
intersection_list = list()
    
count = 0
for ci in intersection_centroids:
    gg = ox.graph_from_point(ci, distance=plot_radius_meters, simplify=False, truncate_by_edge=True, distance_type='bbox',
                             clean_periphery=True, retain_all=False)
    rds = set([e[2].get('name') for e in gg.edges(data=True)])
    if None in rds:
        rds.remove(None)

    if len(rds) == 0:
        continue

    rds = [clean_name(r) for r in rds]
    if len(rds) > 5:
        rd_names = "{}, {} & {}".format(", ".join(rds[:3]), rds[-1], "and more")
    elif len(rds) > 1:
        rd_names = "{} & {}".format(", ".join(rds[:-1]), rds[-1])
    else:
        rd_names = rds[0]

    display(HTML(f"<a href='http://maps.google.com/maps?q={ci[0]},{ci[1]}'>{rd_names}</a>"))

    fname = f"{rds[-1].replace(' ', '_')}_{count}" 
    intersection_list.append((fname, rd_names))

    ox.plot_graph(
        gg, 
        fig_height=5, fig_width=5, bgcolor="#55555500", 
        node_size=0, edge_linewidth=6, edge_color="#1AD8C1",
        use_geom=True, 
        save=True, filename=fname
    )

    count += 1
    if count >= 12:
        break
    
os.chdir(cwd)
````

````python
# Render a pretty picture, inspired by https://www.etsy.com/shop/BarelyMaps 

# First, let's render an html page to do the styling
args = {"intersection_list": intersection_list,
        "location_name": location_name}
with open(os.path.join(result_dir, "index.html"), 'wt') as hf:
    hf.write(render_html_template(os.path.join(code_dir, "single_template.html"), args))
````

````python
# Save html page to an image using imgkit
final_img = os.path.join(output_dir, 'awful_intersections.jpeg')
imgkit.from_file(os.path.join(result_dir, 'index.html'), final_img, options = {'format': 'jpeg', 'quality': 98})
print("Full-res image available in /output/awful_intersections.jpeg")

# Show final image
display(Image.open(final_img))
````
